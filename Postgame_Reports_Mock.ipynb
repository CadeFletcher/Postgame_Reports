```python
import pandas as pd

# Function to convert degrees to clock face with 15-minute intervals
def degrees_to_clock_face_15min(degree):
    hour_segments = ['6:00', '6:15', '6:30', '6:45',
                     '7:00', '7:15', '7:30', '7:45',
                     '8:00', '8:15', '8:30', '8:45',
                     '9:00', '9:15', '9:30', '9:45',
                     '10:00', '10:15', '10:30', '10:45',
                     '11:00', '11:15', '11:30', '11:45',
                     '12:00', '12:15', '12:30', '12:45',
                     '1:00', '1:15', '1:30', '1:45',
                     '2:00', '2:15', '2:30', '2:45',
                     '3:00', '3:15', '3:30', '3:45',
                     '4:00', '4:15', '4:30', '4:45',
                     '5:00', '5:15', '5:30', '5:45']

    # Normalize the degree to be within 0-360 range
    degree = degree % 360

    # Each hour has 30 degrees, each 15 minutes is 7.5 degrees
    # The index in the array can be found by dividing degrees by 7.5
    index = int(degree // 7.5)

    return hour_segments[index]

df['ClockFaceSpinAxis'] = df['ReleaseSpinAxis'].apply(degrees_to_clock_face_15min)

print(df[['ReleaseSpinAxis', 'ClockFaceSpinAxis']].head())
```

```python
import dash
from dash import dcc, html, dash_table
from dash.dependencies import Input, Output
import plotly.graph_objs as go

def calculate_ranges(df, pitcher_id, game_pk):
    pitcher_data = df[(df['PitcherId'] == pitcher_id) & (df['GamePk'] == game_pk)]

    # Perform aggregation
    ranges = pitcher_data.groupby('PitchType').agg(
        Velo_Min=('ReleaseSpeed', 'min'),
        Velo_Max=('ReleaseSpeed', 'max'),
        Velo_Avg=('ReleaseSpeed', 'mean'),
        Horz_Move_Avg=('TrajectoryHorizontalBreak', 'mean'),
        Vert_Move_Avg=('TrajectoryVerticalBreakInduced', 'mean'),
        VAA_Avg=('TrajectoryVerticalApproachAngle', 'mean'),
        HAA_Avg=('TrajectoryHorizontalApproachAngle', 'mean'),
        Spin_Axis=('ClockFaceSpinAxis', lambda x: x.mode()[0] if not x.mode().empty else None),
        SpinRate=('ReleaseSpinRate', 'mean')
    ).reset_index()

    # Apply multiplication after aggregation to convert movement to inches
    ranges['Horz_Move_Avg'] *= 12
    ranges['Vert_Move_Avg'] *= 12

    ranges = ranges.round(1)

    return ranges

def calculate_pitching_stats(df, pitcher_id, game_pk, group_by_pitch_type=True):
    pitcher_data = df[(df['PitcherId'] == pitcher_id) & (df['GamePk'] == game_pk)]

    if group_by_pitch_type:
        group_column = 'PitchType'
    else:
        group_column = 'PitcherId'

    # Define the strike-related pitch calls
    strike_calls = ['called_strike', 'swinging_strike', 'foul', 'foul_bunt', 'foul_tip',
                    'field_out', 'single', 'double', 'home_run', 'grounded_into_double_play',
                    'force_out', 'triple', 'swinging_strike_blocked', 'sac bunt', 'strikeout',
                    'field_error']

    end_of_at_bat_calls = ['strikeout', 'single', 'double', 'triple', 'home_run', 'field_out', 'grounded_into_double_play',
                           'force_out', 'field_error']

    # Create a new column for batters faced
    pitcher_data['Batter_Count'] = pitcher_data['PitchCall'].isin(end_of_at_bat_calls).astype(int)

    # Calculate metrics
    stats = pitcher_data.groupby(group_column).agg(
        Pitches=('PitchCall', 'count'),
        Strikes=('PitchCall', lambda x: (x.isin(strike_calls)).sum()),
        Batters_Faced=('Batter_Count', 'sum'),
        Called_Strike_Rate=('PitchCall', lambda x: (x == 'called_strike').sum() / len(x)),
        Whiff_Rate=('PitchCall', lambda x: (x.isin(['swinging_strike', 'swinging_strike_blocked'])).sum() / len(x)),
        K_Percent=('PitchCall', lambda x: (x == 'strikeout').sum()),
        BB_Percent=('PitchCall', lambda x: (x == 'walk').sum())
    ).reset_index()

    # Post-processing to calculate percentages
    stats['Strike_Percentage'] = (stats['Strikes'] / stats['Pitches']) * 100
    stats['Called_Strike_Rate'] = stats['Called_Strike_Rate'] * 100  # Convert to percentage
    stats['Whiff_Rate'] = stats['Whiff_Rate'] * 100  # Convert to percentage
    stats['CSW_Percent'] = stats['Called_Strike_Rate'] + stats['Whiff_Rate']
    stats['K_Percent'] = (stats['K_Percent'] / stats['Batters_Faced']) * 100
    stats['BB_Percent'] = (stats['BB_Percent'] / stats['Batters_Faced']) * 100
    stats['K_minus_BB'] = stats['K_Percent'] - stats['BB_Percent']

    # Round the results for standard readability
    stats = stats.round(1)

    return stats

color_map = {
    'FF': 'red',    # Fastball
    'SI': 'orange', # Sinker
    'SL': '#FFE600', # Darker yellow for Slider
    'CU': 'blue',   # Curveball
    'CH': 'green',  # Changeup
    'FC': 'brown',  # Cutter
    'KC': 'purple', # Knuckle Curve
}

# Initialize the Dash app
app = dash.Dash(__name__)

# Layout of the app
app.layout = html.Div([
    dcc.Tabs([
        dcc.Tab(label='Strike Zone Report', children=[
            html.Div([
                # Dropdown for game selection
                html.Label('Select Game:'),
                dcc.Dropdown(
                    id='game-dropdown',
                    options=[{'label': game, 'value': game} for game in df['GamePk'].unique()],
                    value=df['GamePk'].unique()[0]
                ),
                # Dropdown for pitcher selection
                html.Label('Select Pitcher:'),
                dcc.Dropdown(
                    id='pitcher-dropdown',
                    value=None
                ),
            ]),
            # Graph for strike zone report
            dcc.Graph(id='pitch-graph')
        ]),
        dcc.Tab(label='Pitch Performance', children=[
            html.Div([
                html.H3("Pitcher Velocity, Movement Ranges, and Angles"),

                # Dropdown for game selection
                html.Label('Select Game:'),
                dcc.Dropdown(
                    id='range-game-dropdown',
                    options=[{'label': game, 'value': game} for game in df['GamePk'].unique()],
                    value=df['GamePk'].unique()[0]
                ),

                # Dropdown to select a pitcher
                html.Label('Select Pitcher:'),
                dcc.Dropdown(
                    id='range-pitcher-dropdown',
                    value=None
                ),

                # DataTable to display velocity and movement ranges
                dash_table.DataTable(
                    id='range-table',
                    columns=[
                        {'name': 'Pitch Type', 'id': 'PitchType'},
                        {'name': 'Velo Min', 'id': 'Velo_Min'},
                        {'name': 'Velo Max', 'id': 'Velo_Max'},
                        {'name': 'Velo Avg', 'id': 'Velo_Avg'},
                        {'name': 'Horz Move Avg (inches)', 'id': 'Horz_Move_Avg'},
                        {'name': 'Vert Move Avg (inches)', 'id': 'Vert_Move_Avg'},
                        {'name': 'VAA Avg', 'id': 'VAA_Avg'},
                        {'name': 'HAA Avg', 'id': 'HAA_Avg'},
                        {'name': 'Spin Axis', 'id': 'Spin_Axis'},
                        {'name': 'Spin Rate', 'id': 'SpinRate'}
                    ],
                    data=[]
                )
            ])
        ]),
        dcc.Tab(label='Pitcher Game Stats', children=[
            html.Div([
                html.H3("Pitcher Game Statistics"),

                # Dropdown to select a pitcher
                html.Label('Select Game:'),
                dcc.Dropdown(
                    id='stats-game-dropdown',
                    options=[{'label': game, 'value': game} for game in df['GamePk'].unique()],
                    value=df['GamePk'].unique()[0]  # Default to first game
                ),

                # Dropdown to select a pitcher
                html.Label('Select Pitcher:'),
                dcc.Dropdown(
                    id='stats-pitcher-dropdown',
                    value=None
                ),

                # Dropdown to select stats by pitch type or total
                html.Label('View Stats:'),
                dcc.Dropdown(
                    id='stats-view-dropdown',
                    options=[
                        {'label': 'By Pitch Type', 'value': 'by_pitch_type'},
                        {'label': 'Total', 'value': 'total'}
                    ],
                    value='total'  # Default to total stats
                ),

                # DataTable to display statistics
                dash_table.DataTable(
                    id='stats-table',
                    columns=[
                        {'name': 'Pitch Type', 'id': 'PitchType'},
                        {'name': 'Strike %', 'id': 'Strike_Percentage'},
                        {'name': 'Batters Faced', 'id': 'Batters_Faced'},
                        {'name': 'Called Strike %', 'id': 'Called_Strike_Rate'},
                        {'name': 'Whiff %', 'id': 'Whiff_Rate'},
                        {'name': 'CSW %', 'id': 'CSW_Percent'},
                        {'name': 'K% (of Batters Faced)', 'id': 'K_Percent'},
                        {'name': 'BB% (of Batters Faced)', 'id': 'BB_Percent'},
                        {'name': 'K%-BB%', 'id': 'K_minus_BB'}
                    ],
                    data=[]
                )
            ])
        ])
    ])
])

# Callback to update both pitcher dropdowns based on the selected game in their respective tabs
@app.callback(
    [Output('pitcher-dropdown', 'options'),
     Output('range-pitcher-dropdown', 'options'),
     Output('stats-pitcher-dropdown', 'options')],
    [Input('game-dropdown', 'value'),
     Input('range-game-dropdown', 'value'),
     Input('stats-game-dropdown', 'value')]
)
def update_pitcher_dropdowns(selected_game, selected_range_game, selected_stats_game):
    # Update for strike zone tab pitcher dropdown
    pitchers_for_strike_zone = df[df['GamePk'] == selected_game]['PitcherId'].unique()
    options_for_strike_zone = [{'label': f'Pitcher {pitcher}', 'value': pitcher} for pitcher in pitchers_for_strike_zone]

    # Update for velocity/movement tab pitcher dropdown
    pitchers_for_ranges = df[df['GamePk'] == selected_range_game]['PitcherId'].unique()
    options_for_ranges = [{'label': f'Pitcher {pitcher}', 'value': pitcher} for pitcher in pitchers_for_ranges]

    # Update for stats tab pitcher dropdown
    pitchers_for_stats = df[df['GamePk'] == selected_stats_game]['PitcherId'].unique()
    options_for_stats = [{'label': f'Pitcher {pitcher}', 'value': pitcher} for pitcher in pitchers_for_stats]

    return options_for_strike_zone, options_for_ranges, options_for_stats

# Callback to update the pitch plot based on selected game, pitcher, and pitch types
@app.callback(
    Output('pitch-graph', 'figure'),
    [Input('game-dropdown', 'value'),
     Input('pitcher-dropdown', 'value')]
)
def update_graph(selected_game, selected_pitcher):
    # Filter the dataset for the selected game and pitcher
    filtered_df = df[(df['GamePk'] == selected_game) & (df['PitcherId'] == selected_pitcher)]

    # Define the strike zone dimensions
    strike_zone_top = round(df['StrikeZoneTop'].mean(), 4)
    strike_zone_bottom = round(df['StrikeZoneBottom'].mean(), 4)
    strike_zone_left = -0.7083
    strike_zone_right = 0.7083

    # Create points for the pitches
    data = []
    for pitch_type in filtered_df['PitchType'].unique():
        pitch_data = filtered_df[filtered_df['PitchType'] == pitch_type]
        trace = go.Scatter(
            x=pitch_data['TrajectoryLocationX'],
            y=pitch_data['TrajectoryLocationZ'],
            mode='markers',
            marker=dict(color=color_map.get(pitch_type, 'gray')),
            name=pitch_type
        )
        data.append(trace)

    layout = go.Layout(
        plot_bgcolor='#D3D3D3',
        shapes=[
            dict(
                type='rect',
                x0=strike_zone_left,
                x1=strike_zone_right,
                y0=strike_zone_bottom,
                y1=strike_zone_top,
                line=dict(color='black', width=1.5)
            )
        ],
        xaxis=dict(
            title='Horizontal Location (feet)',
            range=[-2.5, 2.5],
            showgrid=True
        ),
        yaxis=dict(
            title='Vertical Location (feet)',
            range=[0, 5.0],
            showgrid=True
        ),
        title=f'Pitch Locations for Pitcher {selected_pitcher} (Catcher POV)',
        hovermode='closest',
        autosize=False,
        width=600,
        height=700
    )

    return go.Figure(data=data, layout=layout)

# Callback to update the range table based on selected game and pitcher
@app.callback(
    Output('range-table', 'data'),
    [Input('range-game-dropdown', 'value'),
     Input('range-pitcher-dropdown', 'value')]
)
def update_range_table(selected_game, selected_pitcher):
    ranges = calculate_ranges(df, selected_pitcher, selected_game)
    return ranges.to_dict('records')

# Callback to update the stats table based on selected game, pitcher, and stats view option
@app.callback(
    Output('stats-table', 'data'),
    [Input('stats-game-dropdown', 'value'),
     Input('stats-pitcher-dropdown', 'value'),
     Input('stats-view-dropdown', 'value')]
)
def update_stats_table(selected_game, selected_pitcher, view_option):
    if view_option == 'by_pitch_type':
        stats = calculate_pitching_stats(df, selected_pitcher, selected_game, group_by_pitch_type=True)
    else:
        stats = calculate_pitching_stats(df, selected_pitcher, selected_game, group_by_pitch_type=False)

    return stats.to_dict('records')

# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```

```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import dash_table
import plotly.graph_objs as go
import pandas as pd


# Define the pitch type color map
color_map = {
    'FF': 'red',    # Fastball
    'SI': 'orange', # Sinker
    'SL': '#FFE600', # Darker yellow for Slider
    'CU': 'blue',   # Curveball
    'CH': 'green',  # Changeup
    'FC': 'brown',  # Cutter
    'KC': 'purple', # Knuckle Curve
}

# Define swing-related pitch calls
swing_calls = ['swinging_strike', 'foul', 'foul_tip', 'field_out', 'single', 'double',
               'home_run', 'grounded_into_double_play', 'force_out', 'triple', 'swinging_strike_blocked', 'field_error']

# Initialize the Dash app
app = dash.Dash(__name__)

# Layout of the app
app.layout = html.Div([
    dcc.Tabs([
        dcc.Tab(label='Batter Report', children=[
            html.Div([
                html.Label('Select Game:'),
                dcc.Dropdown(
                    id='game-dropdown',
                    options=[{'label': game, 'value': game} for game in df['GamePk'].unique()],
                    value=df['GamePk'].unique()[0]
                ),
                html.Label('Select Batter:'),
                dcc.Dropdown(
                    id='batter-dropdown',
                    value=None
                ),
                html.Label('Select At Bat:'),
                dcc.Dropdown(
                    id='atbat-dropdown',
                    value=None
                ),
                dcc.Graph(id='pitch-graph')
            ])
        ]),
        dcc.Tab(label='Triple Slash Line', children=[
            html.Div(id='triple-slash-container', children=[
                html.Label('Select Total or Game:'),
                dcc.Dropdown(
                    id='total-game-dropdown',
                    options=[
                        {'label': 'Total', 'value': 'Total'}
                    ] + [{'label': game, 'value': game} for game in df['GamePk'].unique()],
                    value='Total'
                ),
                html.Label('Select Batter for Triple Slash:'),
                dcc.Dropdown(
                    id='triple-slash-batter-dropdown',
                    value=None
                ),
                dash_table.DataTable(
                    id='triple-slash-table',
                    columns=[{'name': 'Batter', 'id': 'Batter'},
                             {'name': 'Batting Average', 'id': 'Batting Average'},
                             {'name': 'On-Base Percentage', 'id': 'On-Base Percentage'},
                             {'name': 'Slugging Percentage', 'id': 'Slugging Percentage'},
                             {'name': 'Total PAs', 'id': 'Total PAs'},
                             {'name': 'Total At Bats', 'id': 'Total At Bats'},
                             {'name': 'Total Walks', 'id': 'Total Walks'},
                             {'name': 'Swing Percentage', 'id': 'Swing Percentage'},
                             {'name': 'Chase Percentage', 'id': 'Chase Percentage'}],
                    data=[],
                )
            ])
        ])
    ])
])

# Callback to update the batter dropdown based on the selected game
@app.callback(
    Output('batter-dropdown', 'options'),
    Input('game-dropdown', 'value')
)
def update_batter_dropdown(selected_game):
    batters_for_game = df[df['GamePk'] == selected_game]['BatterId'].unique()
    return [{'label': f'Batter {batter}', 'value': batter} for batter in batters_for_game]

# Callback to update the at-bat dropdown based on the selected game and batter
@app.callback(
    Output('atbat-dropdown', 'options'),
    [Input('game-dropdown', 'value'),
     Input('batter-dropdown', 'value')]
)
def update_atbat_dropdown(selected_game, selected_batter):
    filtered_df = df[(df['GamePk'] == selected_game) & (df['BatterId'] == selected_batter)]
    filtered_df['SequentialAtBat'] = filtered_df.groupby('BatterId')['AtBatNumber'].rank(method='dense').astype(int)
    atbats = filtered_df['SequentialAtBat'].unique()
    return [{'label': f'At-Bat {atbat}', 'value': atbat} for atbat in atbats]

# Callback to update the pitch plot based on selected game, batter, and at-bat
@app.callback(
    Output('pitch-graph', 'figure'),
    [Input('game-dropdown', 'value'),
     Input('batter-dropdown', 'value'),
     Input('atbat-dropdown', 'value')]
)
def update_graph(selected_game, selected_batter, selected_atbat):
    filtered_df = df[(df['GamePk'] == selected_game) & (df['BatterId'] == selected_batter)]
    filtered_df['SequentialAtBat'] = filtered_df.groupby('BatterId')['AtBatNumber'].rank(method='dense').astype(int)
    if selected_atbat is not None:
        filtered_df = filtered_df[filtered_df['SequentialAtBat'] == selected_atbat]

    data = []
    for pitch_type in filtered_df['PitchType'].unique():
        pitch_data = filtered_df[filtered_df['PitchType'] == pitch_type]
        trace = go.Scatter(
            x=pitch_data['TrajectoryLocationX'],
            y=pitch_data['TrajectoryLocationZ'],
            mode='markers',
            marker=dict(color=color_map.get(pitch_type, 'gray')),
            name=pitch_type
        )
        data.append(trace)
        swung_data = pitch_data[pitch_data['PitchCall'].isin(swing_calls)]
        if not swung_data.empty:
            trace_with_outline = go.Scatter(
                x=swung_data['TrajectoryLocationX'],
                y=swung_data['TrajectoryLocationZ'],
                mode='markers',
                marker=dict(color='rgba(0, 0, 0, 0)', line=dict(color='black', width=1.5)),
                name=f'{pitch_type} (Swung)'
            )
            data.append(trace_with_outline)

    layout = go.Layout(plot_bgcolor='#D3D3D3',
        shapes=[
            dict(
                type='rect',
                x0=-0.7083,
                x1=0.7083,
                y0=round(df['StrikeZoneBottom'].mean(), 4),
                y1=round(df['StrikeZoneTop'].mean(), 4),
                line=dict(color='black', width=1.5)
            )
        ],
        xaxis=dict(title='Horizontal Location (feet)', range=[-2.5, 2.5], showgrid=True),
        yaxis=dict(title='Vertical Location (feet)', range=[0, 5.0], showgrid=True),
        title=f'Pitch Locations for Batter {selected_batter} (Catcher POV)',
        hovermode='closest',
        autosize=False,
        width=600,
        height=700,
        showlegend=True
    )

    return go.Figure(data=data, layout=layout)

# Callback to update the triple slash batter dropdown based on the total game selection
@app.callback(
    Output('triple-slash-batter-dropdown', 'options'),
    Input('total-game-dropdown', 'value')
)
def update_triple_slash_batter_dropdown(selected_game_or_total):
    if selected_game_or_total == 'Total':
        batters = df['BatterId'].unique()
    else:
        batters = df[df['GamePk'] == selected_game_or_total]['BatterId'].unique()

    return [{'label': f'Batter {batter}', 'value': batter} for batter in batters]

# Callback to calculate and display the triple slash line
@app.callback(
    Output('triple-slash-table', 'data'),
    [Input('total-game-dropdown', 'value'),
     Input('triple-slash-batter-dropdown', 'value')]
)
def update_triple_slash(selected_game_or_total, selected_batter):
    if selected_batter is None:
        return []

    if selected_game_or_total == 'Total':
        batter_data = df[df['BatterId'] == selected_batter]
    else:
        batter_data = df[(df['BatterId'] == selected_batter) & (df['GamePk'] == selected_game_or_total)]

    # Calculate total hits, total at-bats, and total walks
    total_hits = (batter_data['PitchCall'].isin(['single', 'double', 'triple', 'home_run'])).sum()
    total_PAs = batter_data['AtBatNumber'].nunique()
    total_walks = (batter_data['PitchCall'] == 'walk').sum()
    total_at_bats = total_PAs - total_walks

    # Total bases calculation
    singles = (batter_data['PitchCall'] == 'single').sum()
    doubles = (batter_data['PitchCall'] == 'double').sum()
    triples = (batter_data['PitchCall'] == 'triple').sum()
    home_runs = (batter_data['PitchCall'] == 'home_run').sum()

    total_bases = singles * 1 + doubles * 2 + triples * 3 + home_runs * 4

    # Calculate batting average, on-base percentage, and slugging percentage
    batting_average = total_hits / total_at_bats if total_at_bats > 0 else 0
    on_base_percentage = (total_hits + total_walks) / (total_PAs) if (total_PAs) > 0 else 0
    slugging_percentage = total_bases / total_at_bats if total_at_bats > 0 else 0

    total_swinging_strikes = (batter_data['PitchCall'].isin(['swinging_strike', 'foul', 'foul_tip',
                    'field_out', 'single', 'double', 'home_run', 'grounded_into_double_play',
                    'force_out', 'triple', 'swinging_strike_blocked', 'field_error'])).sum()

     # Total pitches seen using PitchNumber
    total_pitches_seen = batter_data['PitchNumber'].count()

    #Calculate Swing% (swings per total pitches seen)
    swing_percentage = total_swinging_strikes / total_pitches_seen if total_pitches_seen > 0 else 0

    # Calculate Chase%
    strike_zone_bottom = batter_data['StrikeZoneBottom'].mean() if not batter_data['StrikeZoneBottom'].isnull().all() else 0
    strike_zone_top = batter_data['StrikeZoneTop'].mean() if not batter_data['StrikeZoneTop'].isnull().all() else 0

    # Define edges of the plate
    strike_zone_x_left = -0.7083
    strike_zone_x_right = 0.7083

    # Filter for pitches that are outside the strike zone
    outside_strike_zone = batter_data[
        (batter_data['TrajectoryLocationZ'] < strike_zone_bottom) |
        (batter_data['TrajectoryLocationZ'] > strike_zone_top) |
        (batter_data['TrajectoryLocationX'] < strike_zone_x_left) |
        (batter_data['TrajectoryLocationX'] > strike_zone_x_right)
    ]

    total_chase_swinging = (outside_strike_zone['PitchCall'].isin(['swinging_strike', 'foul', 'foul_tip',
                    'field_out', 'single', 'double', 'home_run', 'grounded_into_double_play',
                    'force_out', 'triple', 'swinging_strike_blocked', 'field_error'])).sum()

    total_outside_pitches = outside_strike_zone.shape[0]

    chase_percentage = total_chase_swinging / total_outside_pitches if total_outside_pitches > 0 else 0

    # Return the results as a list of dictionaries for DataTable
    return [{
        'Batter': selected_batter,
        'Batting Average': f"{batting_average:.3f}",
        'On-Base Percentage': f"{on_base_percentage:.3f}",
        'Slugging Percentage': f"{slugging_percentage:.3f}",
        'Total PAs': total_PAs,
        'Total At Bats': total_at_bats,
        'Total Walks': total_walks,
        'Swing Percentage': f"{swing_percentage:.3f}",
        'Chase Percentage': f"{chase_percentage:.3f}",
    }]


# Run the app
if __name__ == '__main__':
    app.run_server(debug=True)
```
